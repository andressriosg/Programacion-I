La palabra reservada *def* se usa para definir funciones. Debe seguirle el nombre de la función y la lista de parámetros
formales entre paréntesis. Las sentencias que forman el cuerpo de la función empiezan en la línea siguiente, y deben estar con sangría.

La primer sentencia del cuerpo de la función puede ser opcionalmente una cadena de texto literal; esta es la cadena de texto de documentación de la función, o docstring.

Hay herramientas que usan las docstrings para producir automáticamente documentación en línea o imprimible, o para
permitirle al usuario que navegue el código en forma interactiva; es una buena práctica incluir docstrings en el código que uno escribe, por lo que se debe hacer un hábito de esto.

```python
def saludar(nombre):
    print("Hola", nombre)
saludar("Andrés")
```

Acá se observa que hay argumentos por defecto dentro de una función. En caso de que no se usen argumentos distintos, entonces toma tales valores.

```python
def operaciones(a, b):
    suma = a + b
    resta = a - b
    return [suma, resta]

s = operaciones(b = 10, a = 5)
print(s)
```

Una función puede tener múltiples salidas

```python
def operaciones(a, b):
    suma = a + b
    resta = a - b
    return (suma, resta)

s, r = operaciones(10, 5)
print(s, r)
```

```python
def fib(n):
    a, b = 0, 1
    while a < n:
        print(a, end=" ")
        a, b = b, a + b

fib(5)
```

```python
import math

def area_tronco_cono(R1, R2, a, h):
    A = math.pi * (R1**2 + R2**2 + a * (R1 + R2))
    V = (h * math.pi / 3) * (R1**2 + R2**2 + R1 * R2)
    return (A,V)

# ejemplo
R1 = float(input("Ingrese el radio 1: "))
R2 = float(input("Ingrese el radio 2: "))
a = float(input("Ingrese la altura: "))
h = float(input("Ingrese la generatriz: "))

print("El área y el volumen son respectivamente: ", area_tronco_cono(R1, R2, a, h))
```

```python
def contador_vocales(palabra):
  vocales = "aeiou"
  consonantes = "bcdfghjklmnñpqrstvwxyz"
  contador = 0
  contador1 = 0

  for letra in palabra.lower():
    if letra in vocales:
      contador += 1
    elif letra in consonantes:
      contador1 += 1
    else:
      continue
  print("Cantidad de vocales:", contador)
  print("Cantidad de consonates:", contador1)
```

```python
contador_vocales("Hola4")
```

```python
pal = str(input("introduce la palabra "))
contador_vocales(pal)
```

```python
def contador_vocales1(palabra):
  vocales = "aeiou"
  consonantes = "bcdfghjklmnñpqrstvwxyz"
  contador = 0
  contador1 = 0
  contador2 = 0

  for letra in palabra.lower():
    if letra in vocales:
      contador += 1
    elif letra in consonantes:
      contador1 += 1
    else:
      contador2 += 1
  return({"Cantidad de vocales:": contador, "Cantidad de consonates:": contador1, "Cantidad de otros:": contador2})
```

```python
def invertir(cadena):
    return cadena[::-1] # cadena[inicio:fin:paso]
invertir("carro")
```

```python
def es_palindromo(cadena):
    cadena = cadena.lower()
    if cadena == cadena[::-1]:
      return "Verdadero"
    else:
      return "Falso"
```

```python
es_palindromo("camara")
```

```python
pal = input(str("introduce la palabra "))
es_palindromo(pal)
```

```python
def buscar(diccionario, clave):
    if clave in diccionario:
        return diccionario[clave]
    else:
        return "No existe"
```

```python
persona = {"nombre": "Ana", "edad": 20}
print(buscar(persona, "nombre"))
print(buscar(persona, "correo"))
```

```python
def buscarclave(diccionario, valor):
  for i in range(len(diccionario)):
    if valor == list(diccionario.items())[0:len(diccionario)][i][1]:
      return list(diccionario.items())[0:len(diccionario)][i][0]
    else:
      return "No existe"
```

```python
buscarclave(persona, "Ana")
```
