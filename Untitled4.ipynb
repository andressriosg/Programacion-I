```python
def reconocerparimpar(m):
  try:
    m = float(m)
    if m.is_integer() and m >= 0:
      for n in range(0, int(m + 1)):
        if n % 2 == 0:
          print(n, "es par y su doble es", n * 2)
        else:
          print(n, "es impar y su triple es", n * 3)
    elif m.is_integer() and m < 0:
        print("El número es negativo")
    else:
        print("El número no es entero")
  except ValueError:
    print("El valor ingresado no es un número")
```

```python
reconocerparimpar(6)
```

```python
def contador_vocales1(palabra):
  letras = "aeioubcdfghjklmnñpqrstvwxyz"
  letras1 = letras.upper()
  numeros = "0123456789"
  espacio = " "

  contador = 0
  contador1 = 0
  contador2 = 0
  contador3 = 0
  contador4 = 0

  for letra in palabra:
    if letra in letras:
      contador += 1
    elif letra in letras1:
      contador1 += 1
    elif letra in numeros:
      contador2 += 1
    elif letra in espacio:
      contador3 += 1
    else:
      contador4 += 1
  return({"Cantidad de letras minusculas:": contador,
          "Cantidad de letras mayusculas:": contador1,
          "Cantidad de números:": contador2,
          "Cantidad de espacios": contador3,
          "Cantidad de otros caracteres": contador4})
```

```python
contador_vocales1("casa123.K ")
```

```python
palabrea = input("Ingrese una cadena: ")
contador_vocales1(palabrea)
```

```python
def producto_cruz(v1, v2):
  if len(v1) != 3 or len(v2) != 3:
    return "Los vectores deben ser de 3 dimensiones"
  else:
    resultado = [
        v1[1]*v2[2] - v1[2]*v2[1],
        v1[2]*v2[0] - v1[0]*v2[2],
        v1[0]*v2[1] - v1[1]*v2[0]
        ]
  return resultado
```

```python
def producto_cruz(v1, v2):
  try:
    v1 = list(map(float, v1))
    v2 = list(map(float, v2))
    if len(v1) != 3 or len(v2) != 3:
      return "Los vectores deben ser de 3 dimensiones"
    else:
      resultado = [
          v1[1]*v2[2] - v1[2]*v2[1],
          v1[2]*v2[0] - v1[0]*v2[2],
          v1[0]*v2[1] - v1[1]*v2[0]
          ]
      return resultado
  except ValueError:
    return "Los vectores deben tener entradas numéricas"
```

```python
v1 = ["a", 2, 3]
v2 = [3.8, 5, 6]
producto_cruz(v1, v2)
```
